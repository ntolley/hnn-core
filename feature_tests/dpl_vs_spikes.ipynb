### Exploring format of dpl and spikes objects

```python
import os.path as op
# # import tempfile
import hnn_core
from hnn_core import JoblibBackend
from hnn_core import simulate_dipole, read_params, Network, read_spikes, viz
import matplotlib.pyplot as plt
hnn_core_root = op.dirname(hnn_core.__file__)

params_fname = op.join(hnn_core_root, 'param', 'default.json')
params = read_params(params_fname)
params['tstop'] = 50

net = Network(params)
```

```python
from hnn_core.tests.test_dipole import test_dipole
ax = plt.figure()
test_dipole()
```

```python
with JoblibBackend(n_jobs=1):
    dpls = simulate_dipole(net, n_trials=3)
```

```python
params['celsius'] = 50
net = Network(params)

with JoblibBackend(n_jobs=1):
    dpl2 = simulate_dipole(net, n_trials=3)
```

```python
viz.plot_dipole([dpls[0],dpl2[0]], layer='L5')
```
